```python
from scipy.stats import zscore
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def npp(data):
    data = sorted(data)
    p = [(data.index(i)-0.5)/len(data) for i in data]
    z = zscore(p)
    
    plt.scatter(data,z)
    plt.ylabel('Z values')
    plt.xlabel('Values')
    plt.show()
    plt.plot(data, z, 'ro', data , data)
    plt.show()
```

```python
npp(df)
```
